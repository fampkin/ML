```python
print(" ПАУТИНА СУДЬБЫ ")

print("Вы - Майлз Моралес. После событий 'Через вселенные' ваши паучьи силы нестабильны.")
inventory = []  # список с методами
used_items = set()  # множество с методами

info_dict = {  # готовый словарь
    "костюм": "Основной паучий костюм с чипом невидимости.",
    "web-шутеры": "Усовершенствованные паутинные шутеры."
}

new_dict = {}  # новый словарь, будет пополняться по ходу сюжета

plot = []  # список для хранения выборов игрока и сюжета

# Цикл while — ввод первого выбора
choice = ""
while choice not in ["1", "2", "3"]:
    print("\nСигнал бедствия! На выбор:")
    print("1 — Помочь Гвен в её вселенной")
    print("2 — Остановить Кингпина в Нью-Йорке")
    print("3 — Исследовать аномалию в Бруклине")
    choice = input("Ваш выбор: ")

plot.append(f"Первый выбор: {choice}")

# --- Развитие сюжета ---
if choice == "1":
    print("\nВы открываете портал во вселенную Гвен. Вы с Гвен против Пятна. Её отец-капитан в опасности.")
    inventory.append("костюм")  # использование метода списка
    print("Вы надели костюм и готовы к битве.")
    plot.append("\nВы открываете портал во вселенную Гвен. Вы с Гвен против Пятна. Её отец-капитан в опасности.\nВы надели костюм и готовы к битве.")

    # Выбор внутри первого пути
    action = ""
    while action not in ["1", "2"]:
        print("\nСпасать капитана или остановить Пятна?")
        print("1 — Спасать капитана")
        print("2 — Остановить Пятна")
        action = input("Ваш выбор: ")

    plot.append(f"\nВыбор в первом пути: {action}")

    if action == "1":
        inventory.append("доверие гвен")
        print("Гвен благодарна, но Пятн сбежал.")
        new_dict["спас_капитана"] = True
        plot.append("Гвен благодарна, но Пятн сбежал.")
    else:
        inventory.append("кристалл пятна")
        print("Пятн повержен, но капитан ранен.")
        new_dict["остановил_пятна"] = True
        plot.append("Пятн повержен, но капитан ранен.")

elif choice == "2":
    print("\nВы возвращаетесь в Нью-Йорк. Кингпин строит мультивсечный портал.")
    inventory.append("web-шутеры")
    print("Web-шутеры заряжены и готовы.")
    plot.append("\nВы возвращаетесь в Нью-Йорк. Кингпин строит мультивсечный портал.\nWeb-шутеры заряжены и готовы.")

    # Выбор внутри второго пути
    action = ""
    while action not in ["1", "2"]:
        print("\nАтаковать в лоб или провести диверсию?")
        print("1 — Атаковать в лоб")
        print("2 — Провести диверсию")
        action = input("Ваш выбор: ")

    plot.append(f"\nВыбор во втором пути: {action}")

    if action == "2":
        inventory.append("чертежи портала")
        print("Портал разрушен изнутри.")
        new_dict["диверсия_успешна"] = True
        plot.append("Портал разрушен изнутри.")
    else:
        print("Вы побеждены. Паук-Нуар спасает вас.")
        inventory.append("шрам")
        new_dict["получил_шрам"] = True
        plot.append("Вы побеждены. Паук-Нуар спасает вас.")

else:  # choice == "3"
    print("\nВы исследуете аномалию в Бруклине. Это... другой Майлз?")
    inventory.append("костюм")
    print("Паучье чутьё напряжено до предела.")
    plot.append("\nВы исследуете аномалию в Бруклине. Это... другой Майлз?\nПаучье чутьё напряжено до предела.")

    # Выбор внутри третьего пути
    action = ""
    while action not in ["1", "2"]:
        print("\nДоверять ему или сражаться?")
        print("1 — Доверять")
        print("2 — Сражаться")
        action = input("Ваш выбор: ")

    plot.append(f"\nВыбор в третьем пути: {action}")

    if action == "1":
        inventory.append("союзник")
        print("Он помогает стабилизировать ваши силы.")
        new_dict["доверил_майлзу42"] = True
        plot.append("Он помогает стабилизировать ваши силы.")
    else:
        inventory.append("часть костюма")
        print("Вы побеждаете, но порождаете новый разрыв.")
        new_dict["сражался_с_майлзом"] = True
        plot.append("Вы побеждаете, но порождаете новый разрыв.")

# Третий этап — кризис сил
print("\nВаши силы выходят из-под контроля! Голоса всех Пауков звучат в голове.")
choice3 = ""
while choice3 not in ["1", "2", "3", "4"]:
    print("1 — Использовать кристалл для контроля")
    print("2 — Построить стабилизатор")
    print("3 — Объединить силы с другим Майлзом")
    print("4 — Отдать силы, чтобы спасти мультивселенную")
    choice3 = input("Ваш выбор: ")

plot.append(f"\nТретий выбор: {choice3}")

if choice3 == "1":
    if "кристалл пятна" in inventory:
        inventory.append("кристалл контроля")
        print("\nВы взяли кристалл для контроля. Он пульсирует тёмной энергией...")
        used_items.add("кристалл контроля")
        plot.append("\nВы взяли кристалл для контроля. Он пульсирует тёмной энергией...")
    else:
        print("\nУ вас нет кристалла для контроля.")
        inventory.append("отчаяние")
        plot.append("\nУ вас нет кристалла для контроля.")

elif choice3 == "2":
    inventory.append("стабилизатор")
    print("\nВы построили стабилизатор. Он гудит, пытаясь сдержать ваши силы.")
    used_items.add("стабилизатор")
    plot.append("\nВы построили стабилизатор. Он гудит, пытаясь сдержать ваши силы.")

elif choice3 == "3":
    if "союзник" in inventory:
        inventory.append("динамическое дуо")
        print("\nВы объединились с другим Майлзом.")
        plot.append("\nВы объединились с другим Майлзом.")
    else:
        print("Невозможно объединиться - союзника нет рядом.")
        inventory.append("одиночество")
        plot.append("Невозможно объединиться - союзника нет рядом.")

else:  # choice3 == "4"
    inventory.append("жертва")
    print("\nВы готовы пожертвовать своими силами.")
    used_items.add("жертва")
    plot.append("\nВы готовы пожертвовать своими силами.")

# Последняя часть — анализ списка с циклом for
print("\nВы проверяете инвентарь перед финальным решением...")
for item in inventory:
    print(" -", item)

plot.append("Проверка инвентаря: " + ", ".join(inventory))

# Финал — зависит от различных выборов
print("\nМультивсечная вибрация достигает пика. Судьба всех реальностей решается сейчас...")

# Четвёртая условная конструкция
if "кристалл контроля" in inventory and new_dict.get("остановил_пятна") is True:
    print("\n КОНЦОВКА 1: ХРАНИТЕЛЬ ")
    print("Кристалл даёт полный контроль. Вы становитесь новым защитником мультивсечной!")
    plot.append("\nКОНЦОВКА 1: ХРАНИТЕЛЬ\nКристалл даёт полный контроль. Вы становитесь новым защитником мультивсечной!")

elif "стабилизатор" in inventory and new_dict.get("диверсия_успешна") is True:
    print("\n КОНЦОВКА 2: ВЕЧНАЯ БИТВА ")
    print("Стабилизатор работает! Но Кингпин возвращается... Битва никогда не закончится.")
    plot.append("\nКОНЦОВКА 2: ВЕЧНАЯ БИТВА\nСтабилизатор работает! Но Кингпин возвращается... Битва никогда не закончится.")

elif "динамическое дуо" in inventory and new_dict.get("доверил_майлзу42") is True:
    print("\n КОНЦОВКА 3: ДИНАМИЧЕСКОЕ ДУО ")
    print("Два Майлза - двойная сила. Вы становитесь легендарной командой мечты!")
    plot.append("\nКОНЦОВКА 3: ДИНАМИЧЕСКОЕ ДУО\nДва Майлза - двойная сила. Вы становитесь легендарной командой мечты!")

elif "жертва" in inventory:
    print("\n КОНЦОВКА 4: ЧЕЛОВЕК ")
    print("Вы жертвуете своими силами. Разрывы мультивсечной заживают. Вы снова просто Майлз.")
    plot.append("\nКОНЦОВКА 4: ЧЕЛОВЕК\nВы жертвуете своими силами. Разрывы мультивсечной заживают. Вы снова просто Майлз.")

else:
    print("\n КОНЦОВКА 5: НЕИЗВЕСТНОСТЬ ")
    print("Ваши выборы создали непредсказуемую реальность. Будущее туманно...")
    plot.append("\nКОНЦОВКА 5: НЕИЗВЕСТНОСТЬ\nВаши выборы создали непредсказуемую реальность. Будущее туманно...")

print("\nСпасибо за игру!")

with open("plot.txt", "w", encoding="utf-8") as f:
    for entry in plot:
        f.write(entry + "\n")

print("\nСюжет сохранён в 'plot.txt'.")
```

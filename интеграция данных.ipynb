```python

import matplotlib.pyplot as plt
import seaborn as sns

for how in ['left', 'inner', 'right', 'outer']:
    data_with_agents = data_with_orders.merge(agent_data, on='Agent_name', how=how)
    print(f'merge how={how}, shape={data_with_agents.shape}')
    sns.heatmap(data_with_agents.isnull(), yticklabels=False, cbar=False)
    plt.title(f'merge: {how}')
    plt.show()

agent_data_idx = agent_data.set_index('Agent_name')
for how in ['left', 'inner', 'right', 'outer']:
    data_with_agents = data_with_orders.join(agent_data_idx, on='Agent_name', how=how, rsuffix='_agent')
    print(f'join how={how}, shape={data_with_agents.shape}')
    sns.heatmap(data_with_agents.isnull(), yticklabels=False, cbar=False)
    plt.title(f'join: {how}')
    plt.show()

```

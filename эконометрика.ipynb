```python
x, y = 0.5, 0.5
eps = 1e-3
n = 0

while True:
    x_new = 2 * y - y**2
    y_new = x - 0.5 * x**2
    n += 1
    print(n, round(x_new, 6), round(y_new, 6))
    if max(abs(x_new - x), abs(y_new - y)) < eps:
        x, y = x_new, y_new
        break
    x, y = x_new, y_new

print("Ответ:", round(x, 3), round(y, 3))
```

```python
# Задание 2: Метод Ньютона для IRR

def f(r):
    return -1000 + 300/(1+r) + 400/(1+r)**2 + 500/(1+r)**3

def f_prime(r):
    return -300/(1+r)**2 - 800/(1+r)**3 - 1500/(1+r)**4

# Метод Ньютона
r = 0.1
tolerance = 1e-2
max_iter = 100

for i in range(max_iter):
    r_new = r - f(r) / f_prime(r)
    if abs(r_new - r) < tolerance:
        break
    r = r_new

print(f"IRR методом Ньютона: {r:.6f}")
print(f"Значение функции: {f(r):.2e}")
```

```python
# Задание 3: Метод Гаусса-Зейделя для равновесной цены

# Система уравнений:
# Qd = 100 - 2P
# Qs = 50 + 3P
# Условие равновесия: Qd = Qs

# Преобразуем к виду для итераций:
# Из Qd = Qs получаем: 100 - 2P = 50 + 3P
# P = 10 - это точное решение

# Метод Гаусса-Зейделя применяем к переопределенной системе:
# Q = 100 - 2P
# Q = 50 + 3P

P = 10.0  # начальное приближение
tolerance = 1e-3
max_iter = 100

for i in range(max_iter):
    # Гаусс-Зейдель: используем свежее значение P в каждом уравнении
    Q_new = 100 - 2*P  # из уравнения спроса
    P_new = (Q_new - 50) / 3  # из уравнения предложения
    
    if abs(P_new - P) < tolerance:
        break
    P = P_new

print(f"Равновесная цена: {P:.6f}")
print(f"Равновесный объем: {100 - 2*P:.6f}")
print(f"Проверка (Qd - Qs): {(100 - 2*P) - (50 + 3*P):.2e}")
```

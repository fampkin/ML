Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.

```python
def f(n: int, m: float, s: float): 
    return np.diag(np.arange(m, m + n*s, step=s))
```

```python
f(3, 1, 0.5)
```

Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.

```python
a1 = (np.indices((5,5)).sum(axis=0) % 2) * 5
a1
```

```python
B = np.zeros((5,5))
B[::2, ::2] = 5
B[1::2, 1::2] = 5
B
```

```python
a2 = np.triu(np.ones((5,5)))*5
a2
```

```python
np.linalg.det(a1)
```

```python
np.linalg.det(a2)
```

```python
np.linalg.inv(a2)
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("data.csv")
```

```python
df.tail()
```

```python
count_of_rooms = df["Rooms"].value_counts().sort_index()
count_of_rooms
```

```python
count_of_house_year = df["HouseYear"].value_counts().sort_index()
count_of_house_year
```

```python
explode = [0.1 if i == count_of_rooms.idxmax() else 0 for i in count_of_rooms.index]
plt.figure(figsize=(8, 6))
plt.pie(
    count_of_rooms,
    labels = count_of_rooms.index,
    autopct="%.1f%%",
    explode = explode,
)
plt.title("Распределение квартир по числу комнат (%)", fontsize=14)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(
    df["Price"],
    bins=100
)

plt.xlabel("Цена")
plt.ylabel("Количество квартир")
plt.title("Распределение цен")
plt.show()
```

```python
features = ["Rooms", "Square", "HouseFloor", "HouseYear"]

fig, axes = plt.subplots(2, 2, figsize=(20, 10))

for ax, feature in zip(axes.flatten(), features):
    ax.scatter(df[feature], df["Price"], alpha=0.8)
    ax.set_title(f"Price vs {feature}", fontsize=8)
    ax.set_xlabel(feature)
    ax.set_ylabel("Price")
    ax.grid(True, linestyle="--", alpha=0.5)

plt.suptitle("Диаграммы рассеяния", fontsize=25)
plt.tight_layout()
plt.show()
```

```python
df = pd.read_csv("data.csv")
```

```python
plt.figure(figsize=(8, 5))
sns.kdeplot(df["Price"], fill=True, color="skyblue")
plt.xlabel("Price")
plt.ylabel("Плотность")
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Square"], color="lightgreen")
plt.title("Ящиковая диаграмма признака Square", fontsize=14)
plt.xlabel("(м²)")
plt.show()
```

```python
g = sns.PairGrid(df[['Rooms', 'Square', 'HouseFloor', 'HouseYear', 'Price']], 
                 diag_sharey=False)
g.map_diag(plt.hist, bins=30, alpha=0.7, edgecolor='black')
g.map_lower(sns.kdeplot, fill=True, alpha=0.6, cmap='Blues')
g.map_upper(plt.scatter, alpha=0.6, s=10)

plt.suptitle('PairGrid: Попарные отношения признаков недвижимости', y=1.02)

plt.tight_layout()
plt.show()
```

```python

features = ["Rooms", "Square", "HouseFloor", "HouseYear", "Price"]

corr_matrix = df[features].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Тепловая карта корреляции признаков")
plt.show()
```
